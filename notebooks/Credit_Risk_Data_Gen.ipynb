```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Set seed for reproducibility
np.random.seed(42)
random.seed(42)

num_loans = 25000

# 1. Generate Base Loan Dimensions
loan_ids = [f"LN{100000 + i}" for i in range(num_loans)]
customer_ids = [f"CUST{200000 + random.randint(1, 15000)}" for i in range(num_loans)] 

# Financial variables using realistic distributions
funded_amounts = np.random.choice([5000, 10000, 15000, 20000, 35000, 50000, 75000], size=num_loans, p=[0.2, 0.2, 0.2, 0.15, 0.15, 0.07, 0.03])
terms_months = np.random.choice([36, 60], size=num_loans, p=[0.7, 0.3])

# Simulating credit risk relationships (Lower FICO score = Higher Interest Rate & DTI)
fico_scores = np.clip(np.random.normal(680, 50, num_loans).astype(int), 300, 850)
int_rates = np.round(25.0 - (fico_scores - 300) * 0.025 + np.random.normal(0, 1.5, num_loans), 2)
int_rates = np.clip(int_rates, 5.0, 29.99) 

dti_ratios = np.round(np.clip(np.random.normal(0.30, 0.12, num_loans), 0.05, 0.65), 2)

# Loan purposes & grades
purposes = np.random.choice(['Debt Consolidation', 'Credit Card Refinance', 'Home Improvement', 'Small Business', 'Major Purchase'], size=num_loans, p=[0.5, 0.2, 0.15, 0.05, 0.1])
home_ownership = np.random.choice(['MORTGAGE', 'RENT', 'OWN'], size=num_loans, p=[0.5, 0.4, 0.1])

# Generate realistic historical issue dates over the last few years
start_date = datetime(2023, 1, 1)
issue_dates = [start_date + timedelta(days=random.randint(0, 1100)) for _ in range(num_loans)]

# Create the primary dataframe
df_loans = pd.DataFrame({
    'loan_id': loan_ids,
    'customer_id': customer_ids,
    'funded_amount': funded_amounts,
    'term_months': terms_months,
    'int_rate': int_rates,
    'fico_score': fico_scores,
    'dti': dti_ratios,
    'purpose': purposes,
    'home_ownership': home_ownership,
    'issue_date': issue_dates
})

# EXPORT TO CSV
df_loans.to_csv("raw_loan_data.csv", index=False)
print("Data generation complete! 'raw_loan_data.csv' has been created with 25,000 rows.")
```
